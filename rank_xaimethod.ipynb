```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data from the table provided
data = {
    'Properties': ['Fidelity', 'Consistency', 'Selectivity', 'Robustness', 'Representativeness'],
    'First XAI method': ['SHAP', 'SHAP', 'SHAP', 'SHAP', 'L2X'],
    'Second XAI method': ['L2X', 'LIME', 'LIME', 'L2X', 'SHAP'],
    'Third XAI method': ['LIME', 'L2X', 'L2X', 'LIME', 'LIME']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert method names to numerical values for plotting
methods_map = {'SHAP': 1, 'L2X': 2, 'LIME': 3}

# Apply the mapping to the DataFrame
df_mapped = df.replace(methods_map)

# Create a bar chart to visualize the ranking
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the rankings for each property
bar_width = 0.2
index = np.arange(len(df['Properties']))

bar1 = ax.bar(index, df_mapped['First XAI method'], bar_width, label='First XAI method', color='b')
bar2 = ax.bar(index + bar_width, df_mapped['Second XAI method'], bar_width, label='Second XAI method', color='g')
bar3 = ax.bar(index + 2 * bar_width, df_mapped['Third XAI method'], bar_width, label='Third XAI method', color='r')

# Add labels and titles
ax.set_xlabel('Properties')
ax.set_ylabel('Rank (1 = SHAP, 2 = L2X, 3 = LIME)')
ax.set_title('XAI Methods Ranking per Property')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(df['Properties'])
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Create separate bar charts for each property

properties = df['Properties']

# Plot each property in a separate figure
for property in properties:
    fig, ax = plt.subplots(figsize=(6, 4))
    
    # Plot the bars for this property
    ax.bar(['First XAI method', 'Second XAI method', 'Third XAI method'],
           [df_mapped[df['Properties'] == property]['First XAI method'].values[0],
            df_mapped[df['Properties'] == property]['Second XAI method'].values[0],
            df_mapped[df['Properties'] == property]['Third XAI method'].values[0]],
           color=['b', 'g', 'r'])
    
    # Set labels and title
    ax.set_xlabel('XAI Methods')
    ax.set_ylabel('Rank (1 = SHAP, 2 = L2X, 3 = LIME)')
    ax.set_title(f'XAI Methods Ranking for {property}')
    
    plt.tight_layout()
    plt.show()
```

```python
# Update to correctly map the ranks and calculate the scores for each method

# Initialize a dictionary to store the total score for each XAI method
consensus_ranking = {'SHAP': 0, 'LIME': 0, 'L2X': 0}

# Scoring based on rank (1st = 3 points, 2nd = 2 points, 3rd = 1 point)
scoring = {1: 3, 2: 2, 3: 1}

# Calculate the total score for each XAI method across all properties
for property in properties:
    # Get the methods and their ranks for the current property
    methods = ['SHAP', 'L2X', 'LIME']
    ranks = [
        df_mapped[df['Properties'] == property]['First XAI method'].values[0],
        df_mapped[df['Properties'] == property]['Second XAI method'].values[0],
        df_mapped[df['Properties'] == property]['Third XAI method'].values[0]
    ]
    
    # Assign points based on rank
    consensus_ranking['SHAP'] += scoring[ranks[0]] if ranks[0] == 1 else 0
    consensus_ranking['L2X'] += scoring[ranks[1]] if ranks[1] == 2 else 0
    consensus_ranking['LIME'] += scoring[ranks[2]] if ranks[2] == 3 else 0

# Return the total scores and sort
consensus_ranking
```

```python
# Define new colors for the final consensus ranking chart
final_colors = ['#FFCC00', '#FF6699', '#66CCFF']  # Yellow for SHAP, Pink for L2X, Light Blue for LIME

# Create a 3x2 grid for the plots (3 columns, 2 rows) with updated rank labels and colors
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Flatten the axes for easier iteration
axes = axes.flatten()

# Plot each property in the grid
for i, property in enumerate(properties):
    # Extract values for the current property
    methods = ['SHAP', 'L2X', 'LIME']
    values = [
        df_mapped[df['Properties'] == property]['First XAI method'].values[0],
        df_mapped[df['Properties'] == property]['Second XAI method'].values[0],
        df_mapped[df['Properties'] == property]['Third XAI method'].values[0]
    ]
    
    # Plot on the respective subplot with the specified color
    axes[i].barh(methods, values, color=custom_color_map[property])
    axes[i].set_xlabel('Rank')
    axes[i].set_title(f'XAI Methods Ranking for {property}')

# Add the 6th plot for consensus ranking with new colors
methods = ['SHAP', 'L2X', 'LIME']
consensus_values = [consensus_ranking['SHAP'], consensus_ranking['L2X'], consensus_ranking['LIME']]
axes[5].barh(methods, consensus_values, color=final_colors)  # Use the new colors for SHAP, L2X, LIME
axes[5].set_xlabel('Total Points')
axes[5].set_title('Consensus Ranking of XAI Methods')

plt.tight_layout()
plt.show()
```
