```python
def add_id_to_perf(folder_perf, folder_quality):
    perf = os.listdir(folder_perf)
    quality = os.listdir(folder_quality)
    
    df_qual_total = pd.DataFrame()
    df_perf_total = pd.DataFrame()
    
    for i in range(len(perf)):
        file_perf, file_quality = os.path.join(folder_perf, perf[i]), os.path.join(folder_quality, quality[i])
        df_perf, df_quality = pd.read_csv(file_perf), pd.read_csv(file_quality)
        df_quality["Id"] = df_quality["sequence_length"].astype(str) + df_quality['alpha'].astype(str) + df_quality["upper"].astype(str)
        Id = df_quality.Id.unique()
        df_perf = df_perf.iloc[:len(Id)]
        df_perf['Id'] = Id
        df_qual_total, df_perf_total = pd.concat([df_qual_total, df_quality], ignore_index=True), pd.concat([df_perf_total, df_perf], ignore_index = True)
    return df_qual_total, df_perf_total 
```

```python
def min_max_normalize(data):
    min_val = np.min(data)
    max_val = np.max(data)
    return (data - min_val) / (max_val - min_val)
```

```python
dict_df = dict()
df_names = ['df00' + str(i) for i in range(1,5)]
kk=0
for df in df_quality :
    
    df_lime = df.drop(columns="TW").query('explainer == "lime"')
    df_limefinale = df_lime.drop(columns=['Unnamed: 0', 'Unnamed: 0_x', 'explainer', 'model', 'Unnamed: 0_y',
       'RMSE', 'std_RMSE', 'S_score', 'std_S_score', 'MSE', 'std_MSE',
       'dropout', 'activation', 'batch_size', 'alpha_y', 'upper_y'],
                                axis=1)
    col = df_limefinale.columns
    col_mean = [c + '_mean' for c in col]
    col_std = [c + '_std' for c in col]

    mean_lime = df_limefinale.groupby(["nb_layers"]).mean()
    std_lime = df_limefinale.groupby(["nb_layers"]).std()

    col = mean_lime.columns
    col_mean = [c + '_mean' for c in col]
    col_std = [c + '_std' for c in col]

    mean_lime.columns = col_mean
    std_lime.columns = col_std

    lime_mean_std = pd.concat([mean_lime, std_lime], axis=1)
    lime_mean_std['methods'] = 'lime'

    df_shap = df.drop(columns="TW").query("explainer == 'shap'")
    df_shapfinale = df_shap.drop(columns=['Unnamed: 0', 'Unnamed: 0_x', 'explainer', 'model', 'Unnamed: 0_y',
    'RMSE', 'std_RMSE', 'S_score', 'std_S_score', 'MSE', 'std_MSE',
    'dropout', 'activation', 'batch_size', 'alpha_y', 'upper_y'],
                                axis=1)
    mean_shap = df_shapfinale.groupby(["nb_layers"]).mean()
    std_shap = df_shapfinale.groupby(["nb_layers"]).std()

    mean_shap.columns = col_mean
    std_shap.columns = col_std

    shap_mean_std = pd.concat([mean_shap, std_shap], axis=1)
    shap_mean_std['methods'] = 'shap'

    df_l2x = df.drop(columns="TW").query("explainer == 'l2x'")
    df_l2xfinale = df_l2x.drop(columns=['Unnamed: 0', 'Unnamed: 0_x', 'explainer', 'model', 'Unnamed: 0_y',
    'RMSE', 'std_RMSE', 'S_score', 'std_S_score', 'MSE', 'std_MSE',
    'dropout', 'activation', 'batch_size', 'alpha_y', 'upper_y'],
                                axis=1)
    mean_l2x = df_l2xfinale.groupby(["nb_layers"]).mean()
    std_l2x = df_l2xfinale.groupby(["nb_layers"]).std()

    mean_l2x.columns = col_mean
    std_l2x.columns = col_std

    l2x_mean_std = pd.concat([mean_l2x, std_l2x], axis=1)
    l2x_mean_std['methods'] = 'l2x'

    df_synth = pd.concat([lime_mean_std, shap_mean_std], axis = 0)
    df_synth = pd.concat([df_synth, l2x_mean_std], axis = 0)

    col_for_round = df_synth.columns[:-1]
    col_for_round

    for c in col_for_round:
        df_synth[c] = df_synth[c].round(2)
    df_synth

    col_methods = ['Coherence', 'Congruence', 'Instability', 'Selectivity',
                 'Acumen', 'Verm_stability', 'Fidelity', 'Completness', 'Selectivity', 'methods']

    for i in range(12) : 
        df_synth[col[i]] = df_synth[col_mean[i]].astype(str) + "(" + df_synth[col_std[i]].astype(str) + ")"
    df_synth_vf = df_synth[col_methods]

    df_synth_vf['nb_layers'] = df_synth_vf.index
    dict_df[df_names[kk]] = df_synth_vf
    kk = kk+1
```
